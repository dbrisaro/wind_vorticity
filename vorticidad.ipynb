```python
import xarray as xr 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
CCMPv2 = xr.open_dataset('CCMPv2.dv.vor.gaussiangrid.nc').squeeze()
ERAI = xr.open_dataset('ERAI.dv.vor.gaussiangrid.nc').squeeze()
NCEPR1 = xr.open_dataset('NCEPR1.dv.vor.gaussiangrid.nc').squeeze()
CFSR = xr.open_dataset('CFSR.dv.vor.gaussiangrid.nc').squeeze()
```

```python
# selecciono una longitu fija y miro las series de tiempo 

lon0 = -48

CCMPv2_serie = CCMPv2.sel(lon=lon0, method='nearest')
ERAI_serie = ERAI.sel(lon=lon0, method='nearest')
NCEPR1_serie = NCEPR1.sel(lon=lon0, method='nearest')
CFSR_serie = CFSR.sel(lon=lon0, method='nearest')
```

```python
# ploteo

fig = plt.figure(figsize=(7,3))
ax = plt.axes()
CCMPv2_serie.svo.plot(ax=ax, lw=0.5, label='CCMPv2')
ERAI_serie.svo.plot(ax=ax, lw=0.5, label='ERAI')
NCEPR1_serie.svo.plot(ax=ax, lw=0.5, label='NCEPR1')
CFSR_serie.svo.plot(ax=ax, lw=0.5, label='CFSR')
ax.legend(fontsize=8)
ax.tick_params(labelsize=8)
fig.savefig('prueba_vorticidad', bbox_inches='tight', dpi=300)
```
